```python
def model(data):
    from sklearn.linear_model import LinearRegression
    x = data["YEAR"]
    a=np.array(data[data.keys()[0]])[0]
    b=np.array(data[data.keys()[1]])[0]
    x = np.array(x)
    x = x.reshape(-1,1)
# Create y(labels)
    y = data["CASES"]
    y = np.array(y)
    from sklearn.model_selection import train_test_split
    clf = LinearRegression()
    x_train, x_test, y_train, y_test = train_test_split(x ,y , test_size = 0.2)
    clf.fit(x_train,y_train)
    y_preds = clf.predict(x_test)
    brug=[]
    for i in range(2020,2030):
        pred=int((clf.predict((np.array([i])).reshape(-1,1)))[0])
        if pred<0: pred=0
        brug += [[a,b,i,pred]]
    return(brug)   
    clf.score(x_test,y_test)
```

```python
heart_disease = pd.read_csv("final.csv")
df=[["DISEASE","STATE","YEAR","PREDICTION"]]
df+=model(heart_disease.iloc[0:12])
end=13
for k in range (1,11):
    for i in range (1,36):
        df+=model(heart_disease.iloc[end:end+12])
        end+=12
```
